# **GenAI Roadmap for Beginners**

## 1. What is GenAI?
Generative AI is a type of artificial intelligence that creates new content-such as text, images, music, or code-by learning patterns from existing data, mimicking human creativity.

## 2. GenAI Impact Areas
* Customer support
* Content creation
* Education
* Software development

## 3. Is GenAI Successful?
* Does it solve real world problems?
* Is it useful on a daily basis?
* Is it imacting the world economics?
* Is it creating new jobs?
* Is it accessible?

## 4. The Problem
* Very fast progress
* The FOMO merchants
* No single source to learn

## 5. Solution: Mental Model
#### Foundation Models
* Center of GenerativeAI models.
* These are generaized models. eg. LLMs, LMMs
* Trained on huge data and large hardware.
* Foundation generative models can be divided into 2 parts:
  1. User perspective
  2. Builders perspective

## 6. User Perspective vs Builder Perspective

**User Perspective**
* Prompt engineering
* RAG
* AI agents
* Vector databases
* Fine tuning

**Builder Perspective**
* RLHF (change behavior of LLMs)
* Pretraining (training)
* Quantization
* Fine tuning
