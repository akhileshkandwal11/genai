### Creating Tensors (torch.tensor)
PyTorch tensors are created using `torch.tensor()`

In PyTorch, `torch.tensor(7)` and `torch.Tensor(7)` are different.

**1. `torch.tensor(data)`**

  * Purpose is to create a tensor from existing data (Python scalar, list, NumPy array, etc.).
  * Initializes with the exact value(s) you give.

**2. `torch.Tensor(size)`**

  * Purpose is to create an uninitialized tensor of a given shape.
  * If you pass an integer n, it makes a tensor of shape (n,) with uninitialized memory (random garbage values).

| Expression        | Input meaning | Initializes from | Dtype default      | Safety                    |
| ----------------- | ------------- | ---------------- | ------------------ | ------------------------- |
| `torch.tensor(7)` | Data          | Exact value `7`  | inferred (`int64`) | Safe & predictable        |
| `torch.Tensor(7)` | Size          | Uninitialized    | `float32`          | Unsafe unless overwritten |

## **27.0. Difference between Dot Product and Matrix Multiplication**

The dot product and matrix multiplication are related but not the same  they differ in inputs, output shape, and meaning.

**1. Dot Product**

- Applies to: Two vectors of the same length.  

- Operation: Multiply corresponding elements, then sum the results.  

- Output: A single scalar number.  

$$
\text{dot}(a, b) = \sum_{i=1}^n a_i b_i
$$

- Example:

$$
a = [1, 2, 3], \quad b = [4, 5, 6]
$$

$$
\text{dot}(a, b) = (1\times4) + (2\times5) + (3\times6) = 4 + 10 + 18 = 32
$$

- Interpretation: Measures similarity between two vectors (cosine similarity if normalized).


**2. Matrix Multiplication**

- Applies to: Two matrices (or vectors treated as 2D matrices) where the inner dimensions match.  
  If $A$ is shape $(m, n)$ and $B$ is shape $(n, p)$, the result is shape $(m, p)$.  

- Operation: Each element in the result is a dot product between a row of $A$ and a column of $B$.  

- Example:

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},
\quad
B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

$$
A \times B =
\begin{bmatrix}
(1\times5 + 2\times7) & (1\times6 + 2\times8) \\
(3\times5 + 4\times7) & (3\times6 + 4\times8)
\end{bmatrix}
=
\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}
$$

- Interpretation: Combines linear transformations; the dot product is just the building block of this.


> Matrix multiplication is essentially many dot products between rows of the first matrix and columns of the second.


**Key Difference Table**

| Feature        | Dot Product                         | Matrix Multiplication                |
| -------------- | ----------------------------------- | ------------------------------------ |
| **Inputs**     | Two vectors of same length          | Two matrices (inner dims must match) |
| **Output**     | Scalar                              | Matrix                               |
| **Shape rule** | $(n,) \cdot (n,) \to \text{scalar}$ | $(m,n) \times (n,p) \to (m,p)$       |
| **Meaning**    | Similarity measure                  | Composition of transformations       |

## **27.2. Shape rules for (`torch.matmul`)**
Heres the shape rules for `torch.matmul` so we can predict what it will do in every case.

* Think of it as: `torch.matmul` adapts its behavior depending on dimensionality of inputs.

* The only automatic reshaping is for 1D vectors  PyTorch temporarily treats them as row or column vectors to make the multiplication work, but flattens the output back to 1D or scalar.

| Left Shape    | Right Shape   | Internal Interpretation       | Output Shape  | Description                                  |
| ------------- | ------------- | ----------------------------- | ------------- | -------------------------------------------- |
| `(n,)`        | `(n,)`        | `(1, n)`  `(n, 1)`  scalar  | `()`          | Dot product                                  |
| `(n,)`        | `(n, m)`      | `(1, n)`  `(n, m)`           | `(m,)`        | Vector  Matrix                              |
| `(n, m)`      | `(m,)`        | `(n, m)`  `(m, 1)`           | `(n,)`        | Matrix  Vector                              |
| `(n, m)`      | `(m, p)`      | `(n, m)`  `(m, p)`           | `(n, p)`      | Matrix  Matrix                              |
| `(..., n, m)` | `(..., m, p)` | `(..., n, m)`  `(..., m, p)` | `(..., n, p)` | Batch Matrix  Matrix (broadcast over `...`) |


**1. Both are 1D vectors**

* Shapes: `(n,)`  `(n,)`
* Operation: Dot product
* Result shape: `()` (scalar)

````python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
torch.matmul(a, b)   # 1*4 + 2*5 + 3*6 = 32
````

**2. 2D  2D (Matrices)**

* Shapes: `(m, k)`  `(k, n)`
* Operation: Matrix multiplication
* Result shape: `(m, n)`

````python
A = torch.tensor([[1, 2], [3, 4]])   # (2, 2)
B = torch.tensor([[5, 6], [7, 8]])   # (2, 2)
torch.matmul(A, B)                   # (2, 2)
````

**3. 1D  2D (Vector  Matrix)**

* Shapes: `(k,)`  `(k, n)`
* Operation: Treat 1D vector as row vector `(1, k)`
* Result shape: `(n,)` (row vector result, flattened to 1D)

````python
v = torch.tensor([1, 2])         # (2,)
M = torch.tensor([[3, 4], [5, 6]]) # (2, 2)
torch.matmul(v, M)               # (2,)
````

**4. 2D  1D (Matrix  Vector)**

* Shapes: `(m, k)`  `(k,)`
* Operation: Treat 1D vector as column vector `(k, 1)`
* Result shape: `(m,)` (column vector result, flattened to 1D)

````python
M = torch.tensor([[3, 4], [5, 6]]) # (2, 2)
v = torch.tensor([1, 2])           # (2,)
torch.matmul(M, v)                 # (2,)
````

**5. Higher Dimensions (Batch Matmul)**

* Shapes: `(..., m, k`)  `(..., k, n)`
* Extra leading dims (`...`) are treated as batch dimensions and broadcast.
* The multiplication is applied per batch.

````python
A = torch.randn(10, 3, 4)  # 10 batches of (3x4)
B = torch.randn(10, 4, 5)  # 10 batches of (4x5)
torch.matmul(A, B)     # shape (10, 3, 5)
````

## **28. Matrix Multiplication (Part 2): Two Main Rules of Matrix Multiplication**
**`Note:`** Another most common error in PyTorch and deep learning: shape error.

There are two main rules that performing matrix multiplication needs to satisfy:

**1. Inner dimensions must match:**
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

**2. The resulting matrix has shape of the outer dimensions:**
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3,3)`


### **Inner dimensions must match**

````python
# torch.matmul(torch.rand(3, 2), torch.rand(3, 2))
# RuntimeError: mat1 and mat2 shapes cannot be multiplied (3x2 and 3x2)
````

````python
torch.matmul(torch.rand(2, 3), torch.rand(3, 2))
````

````python
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
````

### **The resulting matrix has shape of the outer dimensions**

````python
torch.matmul(torch.rand(3, 10), torch.rand(10, 3)).shape
````

````python
torch.matmul(torch.rand(7, 10), torch.rand(10, 5)).shape
````

### **Transpose: Fix tensor shape error**
* To fix our tensor shape issues, we can manipulate the shape of one of our tensors using transpose.
* A transpose switches the axes or dimensions of a given tensor.

````python
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
````

````python
tensor_B, tensor_B.shape
````

````python
tensor_B.T, tensor_B.T.shape
````

````python
torch.matmul(tensor_A, tensor_B.T).shape
````

````python
# The tensor multiplication works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <-- inner dimensions must match")
print("\nOutput:")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
````

Methods to manipulate our tensors in some way to change there shape or change there dimensions:
1. **Reshaping:** reshapes an input tensor to a defined shape
2. **View:** returns a view of an input tensor of certain shape but keep the same memory as the orignal tensor.
3. **Stacking:** combine multiple tensors on top of each other (vstack) or side by side (hstack).
4. **Squeeze:** removes all `1` dimension from a tensor.
5. **Unsqueeze:** adda a `1` dimension to a target tensor.
6. **Permute:** Return a view of the input with dimensions permuted (swapped) in a certain way.

### **Reshaping**
* Adding an extra dimension.
* Key thing to keep in mind about reshape is dimension has to be compatible with orignal dimensions.
* Reshaping can be done in multiples of orignal tensor shape multiples only.
&ensp;&ensp;&emsp;&emsp;eg.`torch.Size([10]))` -> `(1, 10)` , `(10, 1)` , `(2, 5)` , `(5, 2)`

````python
# Create a tensor
x = torch.arange(1., 11.)
x, x.shape
````

````python
## Incompatible with orignal dimension
# x_reshaped = x.reshape(1, 7)
# x_reshaped, x_reshaped.shape
# RuntimeError: shape '[1, 7]' is invalid for input of size 10
````

````python
x_reshaped = x.reshape(1, 10)
x_reshaped, x_reshaped.shape
````

````python
x_reshaped = x.reshape(10, 1)
x_reshaped, x_reshaped.shape
````

````python
x_reshaped = x.reshape(5, 2)
x_reshaped, x_reshaped.shape
````

````python
x_reshaped = x.reshape(2, 5)
x_reshaped, x_reshaped.shape
````

### **View**
* Changing `z` changes `x` (because a view of a tensor shares the same memory as the orignal input).
* View shares same memory with the orignal tensor.

````python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
````

````python
z = x.view(1, 9)
z, z.shape
````

**Changing z changes x**

````python
# Changing first value of z changes the first value of x as well
z[:, 0] = 5
z, x
````

### **Stacking**
* It takes list of tensors and concatenates sequence of tensors along a new dimension and we define dimension by:
  * dim = 0
  * dim = 1

**Stack tensors on top of each other (dim=0)**

````python
x = torch.arange(1., 11.)
x, x.shape
````

````python
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
````

**Stack tensors side by side each other (dim=1)**

````python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
````

## **33. Squeezing, Unsqueezing and Permuting Tensors**

### **Squeezing**
* Removes all single dimensions from a target tensor.
* When there is dimension mismatch on tensors, we need to squeeze before we put them into some function.

````python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
````

````python
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
````

````python
# squeeze
x_reshaped.squeeze(), x_reshaped.squeeze().shape
````

````python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
````

### **Unsqueezing**
* Adds a single dimension to a target tensor at a specific dimension:
 * dim = 0
 * dim = 1

````python
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
````

### **Permute**
* Rearranges the dimensions of a target tensor in a specified order.
* Commonly used with images.
* Remember permute is a view. View shares a same memory as orignal tensor.

**Permute the orignal tensor to rearrange the axis (or dim) order**

````python
x_orignal = torch.rand(size=(224, 224, 3))    # Image tensor: [height, width, color_channels]
x_permuted = x_orignal.permute(2, 0, 1)   # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_orignal.shape}")
print(f"New shape: {x_permuted.shape}")   # [color_channels, height, width]
````

**Changes in permute changes orignal**

````python
x_orignal[0, 0, 0]
````

````python
x_orignal[0, 0, 0] = 728218
x_orignal[0, 0, 0], x_permuted[0, 0, 0]
````

## **36. PyTorch Reproducibility (Taking the Random Out of Random)**
In short how a neural network learns:

`start with random numbers -> tensor operations -> update random numbers to try and make them better reperesentations of the data -> again -> again -> again...`

To reduce the randomness in neural network and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the randomness.

Extra resources for reproducibility:

https://pytorch.org/docs/stable/notes/randomness.html

https://en.wikipedia.org/wiki/Random_seed

````python
# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
````

### **Random Seed**

````python
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
````
