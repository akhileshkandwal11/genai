# **Generative AI using LangChain**

## **1. What is LangChain?**

* LangChain is an open source framework that helps in building LLM based applications.
* It provides modular components and end-to-end tools that help developers build complex AI applications, such as chatbots, question-answering systems, retrieval-augmented generation (RAG), autonomus agents, and more.

**Features:**
1. Supports all the major LLMs
2. Simplifies developing LLM based applications
3. Integrations available for all major tools
4. Open source/free/actively developed
5. Supports all major GenAI use cases
